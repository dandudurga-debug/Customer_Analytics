```python
import pandas as pd
```

```python
# Read the CSV data as a Pandas DataFrame
df = pd.read_csv('/content/sample_data/Transational_Data/transaction_dataset.csv', low_memory=False)
# Looking at the first five rows
df.head()
```

```python
# Printing the shape of the dataframe
df.shape
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Dropping the null values
df.dropna(inplace=True)
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Printing a concise summary of the DataFrame
df.info()
```

```python
# Converting CustomerID from float to integer
df['CustomerID'] = df['CustomerID'].astype(int)
# Converting InvoiceNo from object to integer
df['InvoiceNo'] = df['InvoiceNo'].astype(int)
# Converting InvoiceDate from object to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
# Converting Country from object to string
df['Country'] = df['Country'].astype(str)
df.head()
```

```python
# Printing a concise summary of the DataFrame
df.info()
```

```python
# Getting a statistical description of the dataset
df.describe()
```

```python
# Getting the positive value
df['AmountSpent'] = df['AmountSpent'].abs()
df.describe()
```

```python
# Saving as a new CSV file
df.to_csv('/content/sample_data/Transational_Data/Updated_Transational_dataset.csv', header=True, index=False)
```
